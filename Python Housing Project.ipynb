```python
# Import necessary libraries
import numpy as np               # Import NumPy for numerical operations
import pandas as pd              # Import Pandas for data handling
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import seaborn as sns            # Import Seaborn for advanced data visualization

# Read the housing data from a CSV file into a Pandas DataFrame
housing_data = pd.read_excel("housing+(1).xlsx")

# Display basic information about the DataFrame.
housing_data.info()

# Display the first 5 rows of the DataFrame to get an view.
housing_data.head()
```

```python
''' We can see that:
•	the dataset contains 20,640 rows and 10 columns.
•	all columns are numerical except ocean_proximity. 
•	there is some missing values in the column total_bedrooms.'''
```

```python
'''Explaining which feature is nominal, ordinal, discrete or continuous.   
1.Nominal Features (Categorical):
ocean_proximity: This feature represents the proximity of the house to the ocean. It is a categorical feature with nominal values.

2.There are no specific ordinal features in the dataset.

3.Discrete Features (Numerical):
housing_median_age: This feature represents the median age of houses. 
total_rooms: Represents the total number of rooms in a block. 
total_bedrooms: Represents the total number of bedrooms in a block. 
population: Represents the total population in a block. 
households: Represents the total number of households in a block. 

4.Continuous Features (Numerical): 
longitude: Represents the longitude of the house location. 
latitude: Represents the latitude of the house location. 
median_income: Represents the median income of households in a block. 
median_house_value: Represents the median house value for households in a block. '''
```

```python
'''Q1). What is the average median income of the data set and check the distribution of data using appropriate plots.
Please explain the distribution of the plot.'''

# Calculating the average median income and round it to 2 decimal places
avg_median_income = round(housing_data['median_income'].mean(), 2)
print(avg_median_income)

# Create a histogram of the 'median_income' and assigning the number of bins as 20, edge color as 'c', and fill color as 'pink'
sns.histplot(data=housing_data, x='median_income', bins=20, ec='c', fc='pink')

# Set the title of the plot, including the calculated average median income
plt.title(f'Distribution of Median Income (Average: {avg_median_income})')

# Set the x-axis label
plt.xlabel('Median Income')

# Set the y-axis label
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
'''The average median income of the dataset is approximately $3.87. But in the histogram, the distribution of median income is 
slightly skewed to the right. 
The skewness to the right suggests that the majority of households have lower median incomes, however, there are still a significant number of
households with relatively higher median incomes, which contribute to the right tail of the distribution.'''
```

```python
'''Q2). Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.'''

# Creating a histogram with a kernel density estimate (KDE) for the 'housing_median_age' column.
sns.histplot(data=housing_data, x='housing_median_age', bins=50, color = 'r', kde=True)

# Here, 'data' specifies the DataFrame, 'x' is the column to plot, 'bins' sets the number of histogram bins, color is red
# and 'kde=True' includes the KDE plot for a smoother representation of the distribution.

# Add title and labels
plt.title('Distribution of Housing Median Age with KDE')
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
'''Above histogram shows that 'housing_median_age' column is not normaly distributed. Instead, it is skewed or biased around 15 to 35 aged houses.
There is long stick at the end indicates that every house which has age 50 or more are included in that group.
This suggests that there is an uneven distribution of housing ages, with some age ranges having more properties than others.'''
```

```python
'''Q3). Show with the help of visualization, how median_income and median_house_values are related?'''

# Creating a scatter plot, 's' controls the size of data points, and 'alpha' controls the transparency
sns.scatterplot(data=housing_data, x='median_income', y='median_house_value', s=20, c='c', alpha=0.4)

# Add title and labels
plt.title('Income Vs House value')
plt.xlabel('Median Income')
plt.ylabel('Median House value')

# Display the plot
plt.show()
```

```python
'''The scatter plot above does not display a clear positive or negative correlation between these two variables. 
Instead, it shows distinct horizontal lines at various values, particularly around 500k, 450k, and 350k, as if a house had a median value
higher than 500k, it was recorded as 500k in the dataset.'''
```

```python
'''Q4). Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available.'''

# Creating a new DataFrame 'cleaned_housing_data' by removing rows with missing values in the 'total_bedrooms' column
cleaned_housing_data = housing_data.dropna(subset = ['total_bedrooms'])       
#'subset' is used in place for just column name to keep intact other data while removing only rows in which data is not available. 

#Checking shape of the dataset after removing the null values.
print(cleaned_housing_data.shape  )            
```

```python
'''Q5). Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.'''

# Calculating mean for the total_bedroom column.
mean_total_bedrooms = round(housing_data['total_bedrooms'].mean(),2)
print(mean_total_bedrooms)

# Fill missing values with the mean using fillna().
data_with_filled_values = housing_data.fillna({'total_bedrooms': mean_total_bedrooms})
print(data_with_filled_values.shape)
```

```python
'''Q6). Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.'''

# Defining a function 'calculate_median' that calculates the median of a any dataset.
def calculate_median(data):
    median = np.median(data)
    return median

# for example: calculating median for housing_median_age column

median_value = calculate_median(housing_data['housing_median_age'])
print(median_value)
```

```python
'''A function named 'calculate_median' is defined above. Input: data - the dataset for which the median is to be calculated and 
output: median - the calculated median value. To call this function, housing_median_age column is used and the result is 29.0'''
```

```python
'''Q10). Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.'''

# Creating a new column 'total_bedroom_size' and initialize it as None
housing_data['total_bedroom_size'] = None

# Loop through the 'total_bedrooms' column and categorize sizes based on values
for j, i in enumerate(housing_data['total_bedrooms']):
    if i <= 10:
        housing_data.loc[j, 'total_bedroom_size'] = 'Small'
    elif 11 <= i <= 1000:
        housing_data.loc[j, 'total_bedroom_size'] = 'Medium'
    else:
        housing_data.loc[j, 'total_bedroom_size'] = 'Large'

# Display the updated DataFrame
print(housing_data.head())
```

```python
'''First a new column is created for 'total_bedroom_size'. Then 'For' loop is used to track the indexes, finally if-elif-else condition is put to 
classify the number of bedrooms into different categories.
As seen in the result total_bedrooms having value above 1000 are succesfully classified as 'Large' size.'''
```
